## Label 2 vs All

```python
def change_label(y,chosen_label):
    y_tmp=[]
    for i in y :
        if i == chosen_label :
            y_tmp.append(1)
        else :
            y_tmp.append(-1)
    return np.array(y_tmp)
```

```python
y_tmp = change_label(data["y"], 2)
print(y_tmp[train_ids])
```

```python
from sklearn.svm import LinearSVC

svm = LinearSVC(tol=1e-3, C=1.0, max_iter=10000)
svm.fit(data["X"][train_ids], data["y"][train_ids])  
```

```python
Best_C = []
for l in range(15):
    score = []
    C = []
    y_tmp = change_label(data["y"],l)
    for i in range(10,255,10):
        svm = LinearSVC(tol=1e-4, C=0.01*i, max_iter=1000)
        svm.fit(data["X"][train_ids], y_tmp[train_ids])
        score.append(svm.score(data["X"][val_ids], y_tmp[val_ids]))
        C.append(0.01*i)
    mplt.figure()
    mplt.plot(C,score)
    mplt.xlabel("C")
    mplt.ylabel("Score en validation")
    mplt.title("Score sur les données de validation en fonction de C en "+str(l)+"vsAll")
    mplt.savefig('bestC_'+str(l)+'vsAll')
    Best_C.append(C[np.argmax(np.array(score))])
    print("Le meilleur C est "+str(C[np.argmax(np.array(score))])+"pour "+str(l)+"vsAll")
```

```python
#Création de tous les SVM avec leur meilleur C et calcul de leur score en train, validation et test.
SVMs = []
score_train = []
score_val = []
score_test = []
for l in range(15) :
    y_tmp = change_label(data["y"],l)
    svm = LinearSVC(tol=1e-4,C=Best_C[l],max_iter=1000)
    svm.fit(data["X"][train_ids], y_tmp[train_ids])
    SVMs.append(svm)
    score_train.append(svm.score(data["X"][train_ids], y_tmp[train_ids]))
    score_val.append(svm.score(data["X"][val_ids], y_tmp[val_ids]))
    score_test.append(svm.score(data["X"][test_ids], y_tmp[test_ids]))
mplt.figure()
mplt.plot(range(15),score_train, 'g',range(15),score_val, 'b',range(15),score_test, 'r')
mplt.xlabel("Label du one vs All")
mplt.ylabel("Score")
mplt.title("Comparaison des score en train, validation et test en fonction du label du one vs All")
mplt.savefig("oneVSall")
```

```python
def score(SVMs,X,y):
    y_pred=[]
    cnt = 0
    for i in range(len(X)) :
        res=[]
        for svm in SVMs :
            res.append(svm.predict(X[i].reshape(1,len(X[i])))[0])
        res = np.array(res)
        y_pred.append(np.argmax(res))
        if y[i] == np.argmax(res) :
            cnt+=1
    return cnt/len(X)
```

```python
#Exemple de prédiction
score(SVMs,data["X"][test_ids],data["y"][test_ids])
```
